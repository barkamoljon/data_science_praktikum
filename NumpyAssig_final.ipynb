```python
# JAVOBNI SHU YERGA YOZING
from numpy import load
def load_data(file_name):
  """Har bir faylni yuklab ularni ikkita mos massivga ajratuvchi funksiyani tuzish """
  data = np.load(file_name)
  data = data['models'], data['qnts']
  return data
```
